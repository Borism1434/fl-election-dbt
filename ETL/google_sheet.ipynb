### `google_sheet.ipynb` 

This notebook automates the process of loading data from a Google Sheet into a PostgreSQL database. It authenticates using a service account, reads the spreadsheet into a DataFrame, and uploads the cleaned data directly to a target table. This is useful for syncing Google Sheets used in field operations or outreach tracking into a central database for analysis.

```python
import gspread
import pandas as pd
from oauth2client.service_account import ServiceAccountCredentials

# Step 1: Define the scope and authenticate
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name("service_acct.json", scope)
client = gspread.authorize(creds)

# Step 2: Load data from a sheet (replace with your sheet name)
sheet = client.open("sheetname").sheet1

# Step 3: Convert to DataFrame
df = pd.DataFrame(sheet.get_all_records())

# Step 4 (optional): Print or process
print(df.info())
```
