```python
import os 
import shutil

def merge_txt_files(directory):
    """
    Merge all .txt files within the given directory into a single file
    and saves the output file inside the same directory. 
    """
    try:
        print(f"Processing Directory: {directory}")

        # Validate if the directory exists
        if not os.path.isdir(directory):
            print(f"Error: The provided path is not a valid directory: {directory}")
            return None
        print("Directory Exists! Proceeding...")
        
        # Get the Directory name for naming the output file
        directory_name = os.path.basename(os.path.normpath(directory))
        output_filename = os.path.join(directory, f"{directory_name}_combined_date.txt")

        # Initialize an empty list to store .txt filename
        txt_files = []

        # Loop through files in the directory and collect .txt files
        for file_name in os.listdir(directory): 
            if file_name.endswith(".txt"):
                txt_files.append(file_name) 
                
            # Check if there are any .txt files 
        if not txt_files:
            print({f"No .txt files found in {directory}"})
            return None
        
        print(f"Found {len(txt_files)} .txt files, merging ...")

        # Open the output file in write mode
        with open(output_filename, "w", encoding="utf-8") as outfile:
            for txt_file in txt_files:
                file_path = os.path.join(directory, txt_file)

                # Open each text file and append its content 
                with open(file_path, "r", encoding="utf-8") as infile:
                    for line in infile:
                        outfile.write(line) # write each line
                    # outfile.write("\n") # Add a newline for separation
        return output_filename # return the merged file path  for later use 
    except Exception as e:
        print(f"Error: {e}")
        return None
    
def process_directories(): 
    """ 
    Prompts the user for two directory path in a single input, 
    merges their text files separetely, and moves the final merged files into a 'final' directory. 
    """
    try: 
        # Ask user for both directories in one input (separated by space)
        directory_input = input("Enter the two directory paths (separated by a space): ").strip()
        
        # Split input into two directory paths 
        directories = directory_input.split(maxsplit=1)

        # Ensure exactly two directories are provided
        if len(directories) != 2:
            print("Error: Please provided exactly two directory path separated by a space.")
            return
        
        dir1, dir2 = directories # Extract Directories 
        
        # Merge Files for both directories
        merged_file1 = merge_txt_files(dir1)
        merged_file2 = merge_txt_files(dir2)

        # Create "final" directory if it dosen't exist
        final_dir = os.path.join(os.path.dirname(dir1), "final")
        os.makedirs(final_dir, exist_ok=True)
        # Move merged files to "final" dir if they exist
        if merged_file1:
            shutil.move(merged_file1, os.path.join(final_dir, os.path.basename(merged_file1)))
            print(f"Moved {os.path.basename(merged_file1)} to {final_dir}")
        
        if merged_file2:
            shutil.move(merged_file2, os.path.join(final_dir, os.path.basename(merged_file2)))
            print(f"Moved {os.path.basename(merged_file2)} to {final_dir}")
        print("All processing completed successfully!")
    except Exception as e:
        print(f"Erorr: {e}")
# Run the full process
process_directories()
```

```python
import csv 
import os 


def get_table_columns(table_name): 
    """ Fetches the column names for a given table from PostgreSQL."""
    conn, cur = connect_to_db()

    if not conn:
        print("Database connection failed. Connect fetch table columns.")
        return None, None, None
    
    try: 
        full_table_name = f"voterfile.{table_name}"

        cur.execute(f"""
                    select column_name, data_type, character_maximum_length
                    from information_schema.columns
                    where table_schema = 'voterfile' and table_name = '{table_name}'
                    order by ordinal_position;
                    """)
        columns = []
        date_columns = set()
        int_columns = set()
        char_limits = {}
        
        for column_name, data_type, char_max_length in cur.fetchall():
            columns.append(column_name)
            
            # Identify Date Columns
            if data_type.lower() == "date":  
                date_columns.add(column_name)
            
            # Identify Integer Columns
            if "int" in data_type.lower():
                int_columns.add(column_name)
            
            # Store character limits (for VARCHAR columns)
            if char_max_length is not None:
                char_limits[column_name] = char_max_length  # Store the max length

        if not columns:
            print(f"Error: Table '{full_table_name} does not exist or has no columns.")
            return None, None, None, None
        
        print(f"Retrieved {len(columns)} columns for table '{full_table_name}'.")
        return columns, date_columns, int_columns, char_limits
    
    except Exception as e: 
        print("Error fetching columns: {e}")
        return None, None, None, None
    
    finally:
        cur.close()
        conn.close()


def clean_value(value, column_name, date_columns, int_columns, char_limits):
    """ Cleans a single value by converting '*' and empty strings to None (Null in PostgreSQL). """
    value = value.strip() # remove leading/trailing space 

    # Convert '*' or empty fields to None (Null in PostgreSQL)
    if value in ["", "*"]:
        return None

   # Check if the column is a DATA type
    if column_name in date_columns: # Identify data columns 
        try:
            parts = value.split("/") 
            if len(parts) == 3: # Expected mm/dd/yyyy format
                month, day, year = parts 
                formatted_date = f"{year}-{month.zfill(2)}-{day.zfill(2)}"
                return formatted_date  # Convert to YYYY-MM-DD
            else: 
                return None # Invalid date 
        except:
            return None
        
    # Convert empty strings to None for integer columns
    if column_name in int_columns:
        if not value.isdigit():  # Ensure value is numeric
            return None
        return int(value)  # Convert valid numeric strings to integers
    
    # Handle Character Limit (Trim oversized values)
    if column_name in char_limits:
        max_length = char_limits[column_name]
        if len(value) > max_length:
            print(f"Truncating '{column_name}' from {len(value)} to {max_length} chars: {value}")
            return value[:max_length]  # Trim to max length

    return value


def preprocess_and_save_cleaned_file(file_path, cleaned_file_path, table_columns, date_columns, int_columns, char_limits):
    """ Reads, cleans, and saves a cleaned version for COPY. """

    with open(file_path, "r", encoding="utf-8") as infile, \
        open(cleaned_file_path, "w", encoding="utf-8", newline="") as outfile:
        reader = csv.reader(infile, delimiter="\t") #tab-delimited file 
        writer = csv.writer(outfile, delimiter="\t", quoting=csv.QUOTE_MINIMAL) # Keep tab format

        for row in reader: 
            # Skip malformed rows 
            if len(row) != len(table_columns):
                print(f"Skipping malformed row (expected {len(table_columns)}, got {len(row)}): {row}")
                continue

            # Clean row before writing 
            cleaned_row = [clean_value(value, table_columns[i], date_columns, int_columns, char_limits) for i, value in enumerate(row)]
            
            # Write to cleaned file (Ensure NULL values are explicitly written)
            writer.writerow(["NULL" if v is None else v for v in cleaned_row])

    print(f"Cleaned file saved at: {cleaned_file_path}")
    return cleaned_file_path # Return cleaned file path


def upload_txt_postgres(file_path, table_name = "test"):
    """ Uploads a tab-delimited .txt file into a PostgreSQL table inside the 'voterfile' schema. """

    if not os.path.exists(file_path):
        print(f"Error: The file '{file_path}' does not exist.")
        return
    
   # Fetch table columns & date columns dynamically
    table_columns, date_columns, int_columns, char_limits = get_table_columns(table_name)

    if not table_columns:
        print("Could not retrieve table columns. Aborting upload.")
        return
    
    # Save cleaned data for COPY
    cleaned_file_path = file_path.replace(".txt", "_cleaned.txt")
    cleaned_file_path = preprocess_and_save_cleaned_file(
        file_path, cleaned_file_path, table_columns, date_columns, int_columns, char_limits
    )  

   # Connect to PostgreSQL using the existing function 
    conn, cur = connect_to_db()

    if not conn:
        print("Database connection failed. Aborting upload.")
        return 
    
    try:
        print(f"Connected to PostgreSQL. Uploading '{file_path}' to 'voterfile.{table_name}'.")

        # Ensure table name uses correct schema 
        full_table_name = f"voterfile.{table_name}"

        copy_query = f"""
            COPY {full_table_name} from STDIN
            with (format csv, delimiter E'\t', NULL 'NULL');
        """

        with open(cleaned_file_path, "r", encoding="utf-8") as f:
            cur.copy_expert(copy_query, f)

        conn.commit()
        print(f"Successfully uploaded data from '{cleaned_file_path}' using COPY.")

    except Exception as e:
        print(f"Error uploading file: {e}")
    
    finally:
        cur.close()
        conn.close()
        print("Database connection closed.")

file_path = "/Users/borismartinez/Documents/final/RecapDetail_combined_date.txt"
upload_txt_postgres(file_path)
```
